```python
# Example of enumerate function
fruits = ['apple', 'banana', 'cherry', 'date']

# enumerate returns both index and value
for index, fruit in enumerate(fruits):
    print(f"Index {index}: {fruit}")

# You can also start from a different index
print("\nStarting from index 1:")
for index, fruit in enumerate(fruits, start=1):
    print(f"Index {index}: {fruit}")
```

```python
# map(function, iterable)

# Using existing variable `token` from previous cell
sample_tokens = token[:12]
print("Original:", sample_tokens)

# Example 1: normalize words to lowercase
lower_tokens = list(map(str.lower, sample_tokens))
print("Lowercase:", lower_tokens)

# Example 2: get length of each word
lengths = list(map(len, sample_tokens))
print("Lengths:", lengths)

# Example 3: custom function (add punctuation)
def add_exclamation(word):
    return word + "!"

excited_tokens = list(map(add_exclamation, sample_tokens))
print("Custom transform:", excited_tokens)
```
