## Underfitting (High Bias)

Occurs when the model is too simple to capture the underlying patterns in the data.

High error on both training and testing set.

Analogy: Studying for an exam but only skimming the chapter titles. You won't do well on the test or even the practice questions.

### Techniques to Handle Underfitting

1. Increase Model Complexity.

2. Feature Engineering

3. Reduce Regularization (Lower the value of regularization parameter)

## Overfitting (High Variance)

Occurs when the model is too complex and learns the noise in the training data rather than the signal.

Low error on the training set but high error on the testing set.

Analogy: Memorizing the practice questions for an exam verbatim. You score 100% on the practice test, but fail the real exam because the questions are slightly different.

### Techniques to Handle Overfitting

1. Get more data

2. Feature Selection (remove irrelevant or redundant features)

3. Cross-Validation (K-Fold Cross Validation)

4. Regularization (Lasso & Ridge Regression) (Assignment: Explore L1 & L2 Regularization)

5. Early Stopping

# Bias-Variance Tradeoff

Ideally, we want a model with low bias and low variance.

Bias: Error due to erroneous or overly simplistic assumptions in the learning algorithm. High bias causes underfitting.

Variance: Error due to too much complexity in the learning algorithm. High variance causes overfitting.
