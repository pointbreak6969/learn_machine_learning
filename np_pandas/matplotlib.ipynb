```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y, marker='o')
# to change its axes, labels, title, etc.
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Plot')
plt.figure(figsize=(8, 6))
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
# multiple plots using fig, ax 
a = [1, 2, 3, 4, 5]
b = [1, 4, 9, 16, 25]
ax.plot(a, b, marker='s', color='r', label='Squares')
ax.plot(x, y, marker='o') 
print(fig, ax)
plt.show()
```

```python
titanic_data = pd.read_csv('./data/titanic.csv')
titanic_data.head()
```

```python
# print("Shape of dataset", titanic_data.shape)
# print("Info of dataset", titanic_data.info())
print("Columns in dataset", titanic_data.columns)
# print("Description of dataset", titanic_data.describe())
print("Missing values in dataset", titanic_data.isnull().sum())
```

```python
survived_counts = titanic_data['Survived'].value_counts()
fig, ax = plt.subplots(figsize=(6, 4))
ax.bar(["Did Not Survive", "Survived"], survived_counts.values)
ax.set_xlabel('Survival Status')
ax.set_ylabel('Number of Passengers')
ax.set_title('Titanic Survival Counts')
ax.set_yticks(np.arange(0, max(survived_counts.values)+100, 50));
```

```python
sibsp_counts = titanic_data['SibSp'].value_counts()
fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(sibsp_counts.index, sibsp_counts.values)
ax.set_xlabel('Number of Siblings/Spouses Aboard')
ax.set_ylabel('Number of Passengers')
plt.show()
```

```python
# Scatter plot
fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(titanic_data["Age"], titanic_data["Fare"], alpha=0.5, color= "green")
ax.set_yticks(np.arange(0, max(titanic_data["Fare"])+50, 50));
```
