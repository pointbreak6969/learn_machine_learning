# Statistical Measures

Common Statistical Measures: Mean, Median, Mode, Variance, Standard Deviation

![](https://media.geeksforgeeks.org/wp-content/uploads/20240619160346/Mean-Variance-and-Standard-Deviation-1.webp)

Mean (Average) : Sum of all values divided by the number of values. It represents the central tendency or the "typical" value.

`Central Limit Theorem`

Median: Middle Value when the data is sorted

Mode: Most frequently occurring value


*Check the formula/intuition behind the mode calculation for discrete and continuous values*

Variance measures how spread out the data points are from the mean. A high variance means data points are widely dispersed. A low variance means they are clustered tightly around the mean. It's the average of the squared differences from the mean. The differences are squared to make them all positive and to penalize larger deviations more heavily.

Standard Deviation is the square root of variance. It's preferred because it is in the same units as the original data, making it more interpretable than variance. It tells the typical distance a data point is from the mean.
