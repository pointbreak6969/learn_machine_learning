# Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

wine = pd.read_csv('./data/wine_data.csv')
```

```python
wine.head()
```

```python
# Shows the number of rows and columns in the dataset
wine.shape
```

```python
# Displays concise summary of the DataFrame, including data types and non-null counts
wine.info()
```

```python
# Provides descriptive statistics for numerical columns in the DataFrame
wine.describe()
```

```python
wine.columns
```

```python
# Counts the occurrences of each unique value in the 'Class' column
wine['Class'].value_counts()
```

```python
# Lists the unique values present in the 'Class' column
wine['Class'].unique()
```

```python
# Checks for missing values in each column
wine.isnull().sum()

# Checks the missing values in the 'Alcohol' feature
# wine['Alcohol'].isnull().sum()
```

```python
# Identifies duplicate rows in the dataset
wine[wine.duplicated()]
```

```python
# Identifies duplicate rows in specific set of columns in the dataset
wine[wine.duplicated(subset=['Alcohol', 'Malic_acid', 'Ash'])]
```

```python
# Removes duplicate rows from the dataset
wine.drop_duplicates(inplace=True) # if the inplace parameter is unspecified, it does not modify the original DataFrame
```

```python
# Removes duplicate rows from specific set of columns in the dataset
# wine.drop_duplicates(subset=['Alcohol', 'Malic_acid', 'Ash'], inplace=True) # if the inplace parameter is unspecified, it does not modify the original DataFrame
wine_unique = wine.drop_duplicates(subset=['Alcohol'])
```

```python
# Computes the pairwise correlation of columns, excluding NA/null values
wine.corr()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(wine.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
