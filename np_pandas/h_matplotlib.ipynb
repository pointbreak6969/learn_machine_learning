## Line Plot

Used for displaying:
- Continuous data (Eg: Time Series Data)
- Trends, patterns


```python
# Simple Line Plot Example
a = [1, 2, 3, 4, 5]
b = [10, 15, 13, 17, 20]

# Create figure and axes
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(a, b, marker='o', linestyle='-', color='red')
ax.set_title('Simple Line Plot')
ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.plot(a, b, marker='o', linestyle='-', color='red')
plt.title('Simple Line Plot')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.show()
```

```python
titanic = pd.read_csv("./data/titanic.csv")
titanic.head()
```

```python
print("Shape of the dataset:", titanic.shape)
print("Columns in the dataset:", titanic.columns)
print("Missing Values in each column:\n", titanic.isnull().sum())
print("Statistical Summary:\n", titanic.describe())
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

survival_counts = titanic['Survived'].value_counts()
# ytick_intervals = [50, 100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600]

# survival_counts
ax.bar(['Did not survive', 'Survived'], survival_counts.values, color=['blue', 'orange'])
ax.set_title('Survival Counts on the Titanic')
ax.set_xlabel('Survival Status')
ax.set_ylabel('Number of Passengers')
# ax.set_yticks(ytick_intervals)


plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10,6))

sibsp_counts = titanic['SibSp'].value_counts()
print(sibsp_counts)

ax.bar(sibsp_counts.index.astype(str), sibsp_counts.values, color='blue');
```

## Scatter Plot

```python
fig, ax = plt.subplots(figsize=(10, 6))

ytick_intervals = np.arange(0, 601, 50)
ax.scatter(titanic['Age'], titanic['Fare'], alpha=0.6, color='green')
ax.set_title('Age vs Fare Scatter Plot')
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
ax.set_yticks(ytick_intervals);
```

```python
# Creating side-by-side plots with 1 row and 2 columns
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Scatter plot on ax1
ax1.scatter(titanic['Age'], titanic['Fare'], alpha=0.6, color='green')
ax1.set_title('Age vs Fare Scatter Plot')
ax1.set_xlabel('Age')
ax1.set_ylabel('Fare')

# Bar Chart on ax2
survival_counts = titanic['Survived'].value_counts()
ax2.bar(['Did not survive', 'Survived'], survival_counts.values, color=['blue', 'orange'])
ax2.set_title('Survival Counts on the Titanic')
ax2.set_xlabel('Survival Status')
ax2.set_ylabel('Number of Passengers');
```

## Histogram

```python
titanic.head()
```

```python
# Creating a histogram of Age distribution

fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(titanic['Age'].dropna(), bins=40, color='Lightgreen', alpha=0.7, edgecolor='black')
ax.set_title('Age Distribution of Titanic Passengers')
ax.set_xlabel('Age')
ax.set_ylabel('Number of Passengers')
ax.grid(axis='y', alpha=0.5);
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(titanic['Fare'].dropna(), bins=40, color='lightgreen', alpha=0.7, edgecolor='black')
ax.set_title('Fare Distribution of Titanic Passengers')
ax.set_xlabel('Fare')
ax.set_ylabel('Number of Passengers')
ax.grid(axis='y', alpha=0.5)
plt.show()
```
