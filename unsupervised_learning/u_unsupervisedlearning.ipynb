# Unsupervised Learning

Unsupervised Learning is a type of machine learning where we train models on data that has no labels, or target variables. Unlike supervised learning, where we know the correct answers (labels), in unsupervised learning, the algorithm must discover patterns, structures, and relationships in the data on its own.

![](https://devopedia.org/images/article/386/2437.1641922762.jpg)

## Part 1: Clustering

Goal: To group similar data points together.

Algorithms: K-Means, Hierarchical Clustering, DBSCAN, Gaussian Mixture Models

Application: Customer Segmentation, Image Segmentation, Document Organization, Gene Sequence Analysis

## Part 2: Dimensionality Reduction

Goal: Reduce the number of features while preserving important information.

Algorithms: Principal Component Analysis (PCA), t-SNE, Autoencoders, Linear Discriminant Analysis (LDA)

Application: Data Visualization (3D -> 2D), Feature extraction, Noise Reduction, Compression
