t-SNE (t-Distributed Stochastic Neighbor Embedding) is s non-linear dimensionality reduction technique primarily used for visualization of high-dimensional data.


When to use t-SNE:
* Visualizing high-dimensional data (images, text embeddings, etc.)
* Exploring Cluster Structure
* Detecting Outlies
* Understanding Data Distribution

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.manifold import TSNE
from sklearn.datasets import load_digits

# The digits dataset consists of 8x8 images of handwritten digits. 
# Each image has 64 features (pixel values) and a target label (the digit it represents).
```

```python
digits = load_digits()
X = digits.data
y = digits.target
images = digits.images
```

#### Before t-SNE: Viewing Raw Data

Each sample is an 8x8 grid of pixel values.

```python
plt.figure(figsize=(10, 6))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(f'Label: {y[i]}')
    plt.axis('off')
plt.tight_layout()
plt.show()
```

#### After t-SNE Visualization

We reduce the 64-dimensional data to 2 dimensions for visualization.

Even though there are raw pixel values, the model learns to separate the data in different numbers.

```python
# Initialize and fit t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

print("t-SNE fitting completed.")
```

```python
X_tsne.shape
```

```python
# Plot the results
plt.figure(figsize=(10, 8))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='tab10', alpha=0.7)
plt.colorbar()
plt.title('t-SNE visualization of digits dataset')
plt.xlabel('t-SNE component 1')
plt.ylabel('t-SNE component 2')
plt.show()
```
