### Derivative Rules:

1. Sum Rule: (f + g)' = f' + g'
2. Product Rule: (f.g)' = f'g + fg'
3. Quotient Rule: (f/g)' = (f'g - fg')/g^2
4. Chain Rule: (f(g(x)))' = f'((g(x)).g'(x))

Loss function L(w): Derivative tells us how to adjust weight w

* if dL/dw > 0 : decreasing w reduces loss
* if dL/dw < 0 : increasing w reduces loss

### Chain Rule: Heart of Backpropagation

Chain rule handles composite functions (functions within functions)

"Rate of change of outside * rate of change of inside"

If y = f(u) and u = g(x), then:

dy/dx = (dy/du) * (du/dx)

y = (3x + 2)^3
u = 3x + 2

then y = u^3

dy/dx = dy/du * du/dx = 3u^2 * 3 = 9 u^2 = 9(3x + 2)^2

if z = f(x, y), x = g(t), y = h(t), then:

dz/dt = (∂z/∂x)(∂x/∂t) + (∂z/∂y)(∂y/∂t) 

![](https://media.springernature.com/lw685/springer-static/image/art%3A10.1038%2Fs41377-024-01590-3/MediaObjects/41377_2024_1590_Fig3_HTML.png)

Input: x

Layer 2: z2 = w2a1 + b2, a2 = f(z2)

# If y represents the actual value
Loss: L = (a2 - y)^2


dL/dw1 = (dL/da2) * (da2/dz2) * (dz2/da1) * (da1/dz1) * (dz1/dw1)


This is called backpropagation.
