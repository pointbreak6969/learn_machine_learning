## Discrete Probability Distribution

#### Bernoulli Distribution

The bernoulli distribution models a single trial with two outcomes: Success (1), and Failure (0).

Eg: Coin Toss results, Email is spam or not, Click or no click


P(X = 1) = p, P(X = 0) = 1-p

where p = probability of success

![](https://media.geeksforgeeks.org/wp-content/uploads/20231031173218/Bernoulli-Distribution-copy.webp)

Used for Binary Classification labels, Logistic Regression Output

#### Binomial Distribution

It models the number of successes in n independent Bernoulli Trials.

Eg: Number of correct predictions in 100 samples, Number of heads in 10 coin tosses


![](https://www.inchcalculator.com/wp-content/uploads/2022/10/binomial-probability-formula.png)

Used for accuracy calculations, voting classifiers, A/B testing

#### Multinomial Distribution (Extension of Binomial Distribution)

Generalization of Binomial Distribution, .i.e. more than two outcomes.

Used in multi-class classification, word counts in NLP



## Continuous Probability Distribution

#### Uniform Distribution


All values in an interval are equally likely.

Eg: Random number between 0 and 1.

![](https://i.ytimg.com/vi/UC-CBUSQXAo/maxresdefault.jpg)

### Normal (Gaussian) Distribution


![](https://i.ytimg.com/vi/5ytum6Dj73Y/maxresdefault.jpg?sqp=-oaymwEmCIAKENAF8quKqQMa8AEB-AH-CYAC0AWKAgwIABABGF4gXiheMA8=&rs=AOn4CLCDfeGLDFXX8nPdQMgVZ4fwXy016w)

Used in Gaussian Noise Model, Loss Functions, Probabilistic Regression

### Exponential Distribution

It models the time until an event occurs.


Eg: Time until system failure, Time between arrivals, Event Modeling, Survival Analysis, Queue Modeling

### Poisson Distribution

It models the number of events in a fixed interval.

Eg: Emails per hour, Requests per second

Distributions in Machine Learning Models:

* Linear Regression - Normal Distribution
* Logistic Regression - Bernoulli Distribution
* Softmax Classifier - Categorical/Discrete
* Naive Bayes - Categorical/Discrete
* HMM (Hidden Markov Models) - Multinomial + Gaussian
