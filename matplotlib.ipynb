```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y, marker='o')
# to change its axes, labels, title, etc.
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Plot')
plt.figure(figsize=(8, 6))
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
# multiple plots using fig, ax 
a = [1, 2, 3, 4, 5]
b = [1, 4, 9, 16, 25]
ax.plot(a, b, marker='s', color='r', label='Squares')
ax.plot(x, y, marker='o') 
print(fig, ax)
plt.show()
```
